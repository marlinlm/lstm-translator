```python

def show_loss(loss_fname, min_steps = 0, max_steps = 0, offset = 0, scaler = 1):
    loss_out_file=loss_fname
    x=[]
    y=[]
    y_1=[]
    y_2=[]
    idx = 0
    dev=[]
    k=400
    k_2=100
    step = 0
    with open(loss_out_file, 'r', encoding='utf-8') as ff:
        for line in ff.readlines():
            step += 1
            
            if min_steps != 0 and min_steps > step:
                continue
          
            if max_steps != 0 and step >= max_steps:
                break
            token = line.split(',')
            y.append(float(token[3][:-1]) * scaler)
            x.append(idx)
            idx += 1
            _y = y[-k:]
            _y_2 = y[-k_2:]
            
#             if len(_y) >= k:
#                 y_1.append(np.median(_y))
#             else:
#                 y_1.append(None)
            
#             if len(_y_2) >= k_2:
#                 dev.append(np.std(_y_2))
#                 y_2.append(np.median(_y_2))
#             else:
#                 y_2.append(None)
    
            y_1.append(np.mean(_y))

            dev.append(np.std(_y_2))
            y_2.append(np.mean(_y_2))
                
    # plt.plot(x,y,'.',label='raw')
    plt.plot(x[offset:],y_2[offset:],'.',label=f'{k_2}-mean')
    plt.plot(x[offset:],y_1[offset:],'.',label=f'{k}-mean')
    plt.xlabel('iter')
    plt.ylabel('losses')
    plt.title('losses')
    plt.legend(loc=9)
    plt.grid(True)
    plt.show()
    print(str(step), str(y_1[-1]), str(y_2[-1]))
#     plt.plot(x,dev,'.',label='dev')
#     plt.xlabel('steps')
#     plt.ylabel('dev')
#     plt.title('dev')
#     plt.legend(loc=9)
#     plt.grid(True)
#     plt.show()
    
    # plt.plot(x,y,'.',label='raw')
    # plt.xlabel('iter')
    # plt.ylabel('losses')
    # plt.title('losses')
    # plt.legend(loc=9)
    # plt.grid(True)
    # plt.show()
    return step

start = 0
cast_start = 0
step = show_loss("../loss.log",start, 0, cast_start, 0.25)

show_loss("../loss.adam",start, step, cast_start)
show_loss("../loss.log_11",start, step, cast_start)
# show_loss("../loss.log_2",5000)
```
